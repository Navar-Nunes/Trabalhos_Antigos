```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Inicia a contagem de tempo
tic = time.time()

tipo_metodo = 4  # O tipo de método a ser utilizado na execução atual.

# 1 = Forward Time
# 2 = Lax Friedrichs
# 3 = Lax Wendroff
# 4 = Bean Warming

# Declaração das variáveis principais
t_atual = 0
t_delta = 1
t_final = 100

Lx_ini = 1
Lx_final = 500

Lx_delta = 1

total_passos = int((Lx_final - Lx_ini) / Lx_delta) + 1

passos_tempo = np.arange(t_atual, t_final + t_delta, t_delta)
passos_Lx = np.arange(Lx_ini, Lx_final + Lx_delta, Lx_delta)

timerCode = 0
timerGraph = 0

# Valores iniciais físicos
cIni = 0.8
cInj = 0

Qn = np.zeros(total_passos + 1)
Qn[0] = cIni

QiAnt = np.zeros(total_passos + 1)
QiAnt[0] = cIni

QiPos = np.zeros(total_passos + 1)
QiPos[0] = cIni

iContorno = 100
decaimento = 0
u = 0.01
Dif = 0.0002

concAtual = cIni
i = 1

if iContorno > i:
    while i < iContorno:
        i = i + 1
        concAtual = concAtual - decaimento
        if concAtual >= 0:
            Qn[i] = concAtual
            QiAnt[i] = concAtual
            QiPos[i] = concAtual

# Início do método
while t_atual < t_final:
    t_atual = t_atual + t_delta
    i = 2
    while i < total_passos:
        # Qn representa os valores no tempo atual, e QiAnt representa os valores no passo de tempo anterior.
        # Forward Time
        if tipo_metodo == 1:
            Qn[i + 1] = Qn[i] - (u * t_delta / Lx_delta) * (Qn[i] - QiAnt[i])
        # Lax Friedrichs
        elif tipo_metodo == 2:
            Qn[i + 1] = (QiPos[i] + QiAnt[i]) / 2 - (u * t_delta / Lx_delta) * (QiPos[i] - QiAnt[i])
        # Lax Wendroff
        elif tipo_metodo == 3:
            Qn[i + 1] = Qn[i] - (u * t_delta / Lx_delta) * (QiPos[i] - QiAnt[i]) + (
                    (u ** 2) * (t_delta ** 2) / (2 * (Lx_delta ** 2))) * (QiPos[i] - 2 * Qn[i] + QiAnt[i])
        # Bean Warming
        elif tipo_metodo == 4:
            Qn[i + 1] = Qn[i] - (u * t_delta / (2 * Lx_delta)) * (
                    3 * Qn[i] - 4 * QiAnt[i] + QiAnt[i - 1]) + (
                                    (u ** 2) * (t_delta ** 2) / (2 * (Lx_delta ** 2))) * (
                                    Qn[i] - 2 * QiAnt[i] + QiAnt[i - 1])
        QiPos[i + 1] = Qn[i + 1]
        i = i + 1

    j = 1
    while j < total_passos:
        # Qn representa os valores no tempo atual, e QiAnt representa os valores no passo de tempo anterior.
        QAnt[j] = Qn[j]
        j = j + 1

# Salva os dados em arquivos
np.savetxt('Caminho_da_pasta/QAtual.dat', Qn)
np.savetxt('Caminho_da_pasta/QAntigo.dat', QiAnt)

timerCode = time.time() - tic  # finaliza a contagem de tempo

# Gera o gráfico
plt.plot(passos_Lx, Qn, 'ro')
plt.show()

timerGraph = time.time() - tic  # finaliza a contagem de tempo
```
