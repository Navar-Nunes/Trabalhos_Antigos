```python
import numpy as np
import matplotlib.pyplot as plt
import time
import math


# Inicia a contagem de tempo
start_time = time.time()

# Declarao das variveis principais
t_inicio = 0
t_delta = 1
t_final = 60

x_inicio = 0
x_delta = 1
x_final = 30

n = 1    # Valor inicial de "n" no somatrio
n_max = 100  # Valor mximo de "n" no somatrio

delta_t_graph = 5;
delta_x_graph = 10;
t_graph = 5;
x_graph = 0;

# Matriz de x e t
# Lembrando que, em Python, os vetores e matrizes comeam com zero invs de 1
# Temos uma matriz com "t" linhas e "x" colunas
matrix_main = np.zeros((t_final+1, x_final+1))
valores_x = np.zeros(x_final+1)
valores_t = np.zeros(t_final+1)
vetor_x = np.arange(x_final+1)
vetor_t = np.arange(t_final+1)

#print(matrix_main)
#print(valores_x)


# Valores iniciais fsicos

Cn = 0

alpha1 = 2
L = x_final

#matriz_u = np.zeros((len(vetor_x), len(vetor_t)))

# while error > 1:
#    teste

# Preenchimento das condies iniciais

i = 0

for i in range(x_final):
  if i <= 10:
    matrix_main[0][i] = (i-1)/10
  else:
    matrix_main[0][i] = (x_final - (i-1))/20

t = 1
x = 0

# Clculo geral da matriz principal
for t in range(t_final):
  for x in range(x_final):
    while n <= n_max:
      Cn = (9/((math.pow(n,2))*(math.pow((math.pi),2))))*(math.sin((n*(math.pi))/3))

      matrix_main[t][x] = matrix_main[t][x] + (Cn*(math.sin((n*(math.pi)*x)/L))*(math.cos((2*n*(math.pi)*t)/L)))

      n = n+1


    n = 1



i = 0;


# Grfico de u vs x para vrios tempos fixos
#tempos_fixos = [10, 20, 30]  # Exemplo de tempos fixos

plt.figure(figsize=(20, 6))
plt.subplot(1, 2, 1)

while t_graph <= t_final:

  #ax = plt.subplot(111) varivel desnecessria nessa verso

  valores_x = matrix_main[t_graph,:]

  plt.plot(vetor_x, valores_x, label=f'u vs x (t={t_graph})')
  t_graph = t_graph + delta_t_graph

plt.xlabel('Posio (x)')
plt.ylabel('u')
plt.title('Grfico de u vs x para vrios tempos fixos')
plt.xlim([0, x_final])
plt.ylim([-1, 1])
plt.grid(True)
plt.legend()

# Move a legenda para fora do grfico
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))


# Grfico de u vs t para vrias posies fixas de x
# posicoes_x_fixas = [5, 15, 25]  # Exemplo de posies fixas de x

plt.figure(figsize=(20, 6))
plt.subplot(1, 2, 1)

while x_graph <= x_final:

  #ax = plt.subplot(111) varivel desnecessria nessa verso

  valores_t = matrix_main[:,x_graph]

  plt.plot(vetor_t, valores_t, label=f'u vs t (x={x_graph})')
  x_graph = x_graph + delta_x_graph

plt.xlabel('Tempo (t)')
plt.ylabel('u')
plt.title('Grfico de u vs t para vrias posies fixas de x')
plt.xlim([0, t_final])
plt.ylim([-1, 1])
plt.grid(True)
plt.legend()

# Move a legenda para fora do grfico
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Exibir grficos
plt.show()

# Tempo de execuo do cdigo
timer_code = time.time() - start_time

print('\n\nTempo total de execuo: ' ,timer_code)
```
